# Creating Arrays

```python
array_zeros = np.zeros(5)
array_ones = np.ones(5)
array_consts = np.full(5, 2.5)
array_zeros,array_ones,array_consts
```

```python
a = np.array([1, 3, 4, 5])
a
```

**Ranges**:
create an array with a range of elements

```python
np.arange(4)
```

```python
# arange(first number, last number, step size)
np.arange(2,9,2)
```
